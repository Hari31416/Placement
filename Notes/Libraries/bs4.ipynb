```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.crummy.com/software/BeautifulSoup/bs4/doc/"
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
tag = soup.find('a', string='Navigating the tree')
print(tag)
```

```python
for p in tag.parents:
    print(p.name)
    print(p.attrs)
```

```python
for a in soup.find_all('a', string=re.compile('Navigating')):
    print(a["href"])
```

```python
def save_image(url, filename):
    r = requests.get(url)
    with open(filename, "wb") as f:
        f.write(r.content)
```

```python
def clean_name(name):
    return name.replace(" ", "_").replace(":", "").lower()
```

```python
def get_image(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    img_url_raw = soup(id="comic")[0].img["srcset"]
    img_url = img_url_raw.split(" ")[0].split("//")[-1]
    name = soup.find("div", id="ctitle").text
    save_image("https://" + img_url, name+".png")
    # return img_url
```

```python
get_image("https://xkcd.com/")
```
